```python
import numpy as np
from astropy.time import Time
import pandas
import os 
import matplotlib.pyplot as plt
```

```python
#now, we open our csv file. This is important so we can look at the different headers for later on in the code.
```

```python
df = pandas.read_csv('data.csv')
```

```python
df.head()
```

```python
#now, we take a closer look at our x values, which are in mdj(Modified Julian Date.)
```

```python
print(df['mjd'])
```

```python
#Now, we are plotting the graph. This graph represents the light curves from the star. 
#we use plt here. We set the figure size, the x and y axis values, and the plot title. 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (20, 10))
plt.scatter(df['mjd'], df["flux"])
plt.title("S3TIC10000026150")
plt.xlabel("MJD(Modified Julian Date)")
plt.ylabel("Flux")
plt.grid(True)

plt.show()
```

```python
#However, it would be easier for the user if we can convert the mjd x axis values into UTC. We convert all of the values in the header into UTC. 
```

```python

date_array = np.array(df['mjd'])
time_array = Time(date_array, format = 'mjd')

hourangle = time_array.to_value('datetime') 

for x in hourangle: 
    print(x)
```

```python
#now, we plot the graph using the new x values, it is easier for the user to read. We also put it on a logarithmic scale.
```

```python


plt.figure(figsize = (20, 10))

#plt.scatter(hourangle, df["flux"])
#plt.title("light curve plot")

#plt.errorbar(hourangle[100:110], df["flux"][100:110], yerr=df["fluxerr"][100:110], fmt='o')
plt.errorbar(hourangle, df["flux"], yerr=df["fluxerr"], fmt='o')
plt.yscale("log")
plt.xlabel("UTC")
plt.ylabel("Flux")

plt.show()
```

```python
#We can add some error bars to make the graph easier to percieve. This region of the graph is "zoomed in". 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (20, 10))
#plt.scatter(hourangle, df["flux"])
#plt.title("light curve plot")
plt.errorbar(hourangle[100:110], df["flux"][100:110], yerr=df["fluxerr"][100:110],   fmt='o')
plt.xlabel("MJD(Modified Julian Date)")
plt.ylabel("Flux")
plt.show()
```

```python
#Finally, we convert the y axis of the graph into logarithmic values.  
```
